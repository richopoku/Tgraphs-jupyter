```python
df = pd.read_excel('Fnt-FDE-data.xlsx')
df.head()
Noofwatermol = df.Noofwatermol
#print(Noofwatermol)
Active6nots = df.Active6nots
#print(Active6nots)
Active6s = df.Active6s
#print(Active6s)
active6intermediate = df.active6intermediate
#print(active6intermediate)
Noactiveintermediate = df.Noactiveintermediate
#print(Noactiveintermediate)
Noinactivenots = df.Noinactivenots
#print(Noinactivenots)
No6inactives = df.No6inactives
#print(No6inactives)
supermolecule = df.supermolecule
#print(supermolecule)
```

```python
plt.plot(Noofwatermol, Active6nots, color="blue", marker='o', linestyle='solid',label='[Cl$^{-}$@(H$_{2}$O)$_{50}$]', linewidth=2)
plt.plot(Noofwatermol, Active6s, color="blue", marker='o', linestyle='dotted',label='[Cl$^{-}$@(H$_{2}$O)$_{50}$]', linewidth=2)
plt.plot(Noofwatermol, active6intermediate, color="blue", marker='o', linestyle='-.',label='[Cl$^{-}$@(H$_{2}$O)$_{50}$]', linewidth=2)
plt.plot(Noofwatermol, Noactiveintermediate, color="green", marker='s', linestyle='-.',label='[Cl(H$_{2}$O)$^{-}_{6}$@(H$_{2}$O)$_{44}$]',linewidth=2)
plt.plot(Noofwatermol, Noinactivenots, color="green", marker='s', linestyle='solid',label='[Cl(H$_{2}$O)$^{-}_{6}$@(H$_{2}$O)$_{44}$]', linewidth=2)
plt.plot(Noofwatermol, No6inactives, color="green", marker='s', linestyle='dotted',label='[Cl(H$_{2}$O)$^{-}_{6}$@(H$_{2}$O)$_{44}$]', linewidth=2)
plt.plot(Noofwatermol, supermolecule, color="red", linestyle='--',label='[Cl(H$_{2}$O)$_{50}$]$^{-}$', linewidth=2)
plt.legend(loc=1, prop={'size': 6}, handlelength=7)
plt.xlabel('number of water molecules', fontsize=12)
plt.ylabel('Electron binding energies [eV]', fontsize=12)
plt.savefig('Fnt-Fde.pdf', bbox_inches='tight', dpi=400)
plt.show()
plt.close()
```
